```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#Carga de datos desde CSV
df = pd.read_csv("Examen/data.csv", delimiter = ";", decimal = ".", encoding ="utf-8")
df.columns= ['date', "unity", 'service', 'sence', 'period', 'sp','so','distance', 'icf']
df['date'] =  pd.to_datetime(df['date']) #Date transformada en tipo datetime64
df['sa'] = abs(df['so']-df['sp'])
df.head(8)
```

```python
salidas = df['so']-df['sp']
salidas_f = []
salidas_s = []
for i in salidas:
    if i <= 0:
        salidas_f.append(i*-1)
    else:
        salidas_f.append(0)
for i in salidas:
    if i >= 0:
        salidas_s.append(i)
    else:
        salidas_s.append(0)
df['sf'] = salidas_f
df['ss'] = salidas_s
# Salidas faltantes y sobrantes
```

```python
#Información del dataframe
df.info()
```

```python
# Segmentar por mes y Unidad de negocio
df['month'] = df['date'].dt.month_name()
cats = [ 'June', 'July', 'August']
df['month'] = df['month'].astype('category', categories=cats, ordered=True)
df.head(8)
```

```python
# Descripción de los ICF
df_icf = df.groupby(['unity', 'month']).describe().icf
df_icf['var'] = 0
df_icf
```

```python
vec_var = []
vec_var.append(0)
for i in range(0,17):   
    vec_var.append(((df_icf['mean'][i+1]-df_icf['mean'][i])/df_icf['mean'][i+1])*100)
    if i%3 == 0:
        vec_var[i]=0
# Calcula el % variación del ICF con respecto al mes anterior     
```

```python
vec_var
```

```python
mean_var = []
mean_var.append(sum(vec_var[0:3])/2)
for i in range(1,6):
    mean_var.append(sum(vec_var[(0+i*3):(3+i*3)])/3)
# Calcula el promedio de la variación de los últimos tres meses
```

```python
mean_var # Promedio de variación de los últimos tres meses
```

```python
summary = pd.DataFrame({'unity': ['U2', 'U3', 'U4', 'U5', 'U6', 'U7']})
summary['mean_var'] = mean_var 
summary
# El variación más alta es U3
```

## U4 Express

```python
is_U4 = df.loc[:, 'unity'] == 'U4'
df_U4 = df.loc[is_U4]
df_U4.head()
```

```python
df_U4.info()
```

# U4 por mes

```python
df_U4_sum = df_U4.groupby(['month']).sum()
df_U4_sum.head(8)
# Sumatoria
```

```python
df_U4_ser = df_U4.groupby(['month']).mean()
df_U4_ser.dropna().head(8)
# Media 
```

```python
# Funcion donde se ingresa una columna de data frame y retorna un vector
def vec_append(x): 
    vec = []
    for i in x:
        vec.append(i)
    return vec
```

```python
df_unity = pd.DataFrame({'month': ['June', 'July', 'August']})
df_unity['icf_mean'] = vec_append(df_U4_ser.icf)
df_unity['dist_mean'] = vec_append(df_U4_ser.distance)
df_unity['sa_sum'] = vec_append(df_U4_sum.sa) 
df_unity['sa_f'] = vec_append(df_U4_sum.sf) 
df_unity['sa_s'] = vec_append(df_U4_sum.ss)
df_unity
```
