```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sys
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df = pd.read_csv("Examen/data.csv", delimiter = ";", decimal = ".", encoding ="utf-8")
df.columns= ['date', "unity", 'service', 'sence', 'period', 'sp','so','distance', 'icf']
```

```python
df['1-icf'] = 1-df['icf']
df['so-sp'] = abs(df['so']-df['sp'])
df['total_dist'] = df['so-sp']*df['distance']  
df['pond'] = df['total_dist']*df['1-icf']
df['dist'] = df['distance']
df.head(8)
```

## Correlación entre distancia y icf por categoría

```python
is_U2 = df.loc[:, 'unity'] == 'U2'
df_U2 = df.loc[is_U2]
df_U2.head(8)
```

# Ponderaciones

```python
def pareto_plot(df, x=None, y=None, title=None, show_pct_y=False, pct_format='{0:.0%}'):
    xlabel = x
    ylabel = y
    tmp = df.sort_values(y, ascending=False)
    x = tmp[x].values
    y = tmp[y].values
    weights = y / y.sum()
    cumsum = weights.cumsum()
    
    fig, ax1 = plt.subplots()
    ax1.bar(x, y)
    ax1.set_xlabel(xlabel)
    ax1.set_ylabel(ylabel)

    ax2 = ax1.twinx()
    ax2.plot(x, cumsum, '-ro', alpha=0.5)
    ax2.set_ylabel('', color='r')
    ax2.tick_params('y', colors='r')
    
    vals = ax2.get_yticks()
    ax2.set_yticklabels(['{:,.2%}'.format(x) for x in vals])

    # hide y-labels on right side
    if not show_pct_y:
        ax2.set_yticks([])
    
    formatted_weights = [pct_format.format(x) for x in cumsum]
    for i, txt in enumerate(formatted_weights):
        ax2.annotate(txt, (x[i], cumsum[i]), fontweight='heavy')    
    
    if title:
        plt.title(title)
        
    plt.figure(figsize=(100,100))
    plt.tight_layout()
    plt.show()
    
# Permite gráficar pareto 
```

```python
is_range_U2 = df_U2.loc[:, 'icf'] <= 0.80 # Rango entre 0 y 0.80
df_U2_range = df_U2.loc[is_range_U2]
df_U2_range = df_U2_range.groupby('distance').mean()
df_U2_range_s = df_U2_range.sort_values('pond', ascending=False)
df_U2_range_s['cumsum'] =(df_U2_range_s['pond'].cumsum())
df_U2_range_s['%']= df_U2_range_s['cumsum']/sum(df_U2_range_s['pond'])
df_U2_range_s.head(10)
```

## Resumen

```python
def counter(x, inf, sup):
    con = 0    
    for i in x:
        if i > inf and i <= sup:
            con +=1                           
    return con
# función que cuenta número de elementos dentro de cierto de rango
```

```python
def return_max(var1,var2):
    vec1 = []
    vec2 = []
    max_vec = 0
    for i, j in zip(var1,var2):
        if i <= 0.8 and max_vec<j:
            max_vec = j         
    return max_vec
# Retorna la distancia máx de los 80/20 casos más relevante

def return_min(var1,var2):
    vec1 = []
    vec2 = []
    min_vec = 100
    for i, j in zip(var1,var2):
        if i <= 0.8 and min_vec>j:
            min_vec = j         
    return min_vec
# Retorna la distancia min de los 80/20 casos más relevante        
```

```python
df_unity = pd.DataFrame({'unity': ['U2', 'U3', 'U4', 'U5', 'U6', 'U7']})
df_unity['n'] = [len(x1), len(x2), len(x3), len(x4), len(x5), len(x6)]
df_unity['r2'] = [r2_U2, r2_U3, r2_U4, r2_U5, r2_U6, r2_U7]
df_unity['[0, 0.80]'] = [counter(y1,-1,0.80), counter(y2,-1,0.80), counter(y3,-1,0.80), counter(y4,-1,0.80),
                    counter(y5,-1,0.80), counter(y6,-1,0.80)]
df_unity['[0, 0.80] %'] = (df_unity['[0, 0.80]']/df_unity['n'])*100
df_unity[']0.80-1.00]'] = [counter(y1,0.8,1), counter(y2,0.8,1), counter(y3,0.8,1), counter(y4,0.8,1),
                    counter(y5,0.8,1), counter(y6,0.8,1)]
df_unity[']0.80-1] %'] = (df_unity[']0.80-1.00]']/df_unity['n'])*100
df_unity['point_80'] = [df_U2_range_s['pond'].mean(), df_U3_range_s['pond'].mean(), df_U4_range_s['pond'].mean(),
                       df_U5_range_s['pond'].mean(), df_U6_range_s['pond'].mean(), df_U7_range_s['pond'].mean()]
df_unity['total_dist_80'] = [df_U2_range_s['total_dist'].mean(), df_U3_range_s['total_dist'].mean(), 
                             df_U4_range_s['total_dist'].mean(), df_U5_range_s['total_dist'].mean(), 
                             df_U6_range_s['total_dist'].mean(), df_U7_range_s['total_dist'].mean()]
df_unity['r2_80']= [r2_U2_80, r2_U3_80, r2_U4_80, r2_U5_80, r2_U6_80, r2_U7_80]
df_unity['max_80'] = [return_max(df_U2_range_s['%'], df_U2_range_s['dist']),
                     return_max(df_U3_range_s['%'], df_U3_range_s['dist']),
                     return_max(df_U4_range_s['%'], df_U4_range_s['dist']),
                     return_max(df_U5_range_s['%'], df_U5_range_s['dist']),
                     return_max(df_U6_range_s['%'], df_U6_range_s['dist']),
                     return_max(df_U7_range_s['%'], df_U7_range_s['dist'])]
df_unity['min_80'] = [return_min(df_U2_range_s['%'], df_U2_range_s['dist']),
                     return_min(df_U3_range_s['%'], df_U3_range_s['dist']),
                     return_min(df_U4_range_s['%'], df_U4_range_s['dist']),
                     return_min(df_U5_range_s['%'], df_U5_range_s['dist']),
                     return_min(df_U6_range_s['%'], df_U6_range_s['dist']),
                     return_min(df_U7_range_s['%'], df_U7_range_s['dist'])]
df_unity
```
